```python
import tensorflow as tf
from tensorflow.keras import layers, models
import numpy as np
```

```python

model = models.Sequential()
model.add(layers.Conv2D (32, kernel_size=(3,3), padding='valid', activation='relu',

input_shape=(28,28,1)))
model.add(layers.Conv2D (32, kernel_size=(3,3), padding='valid', activation='relu'))
model.add(layers.Conv2D (32, kernel_size=(3,3), padding="valid", activation='relu'))
model.add(layers.Flatten())
model.add(layers.Dense (128, activation='relu'))
model.add(layers.Dense (10, activation='softmax'))
```

```python
model.summary()
```

Every layer reduces the spatial size by 2 pixels per convolution (because 3×3 kernel without padding removes 1 pixel from each edge).

Effect

*   Image shrinks → 28 → 26 → 24 → 22

*   Flattened size: smaller (15,488)
**bold text**
*   Parameters: ~2M

```python
model = models.Sequential()
model.add(layers.Conv2D (32, kernel_size=(3,3), padding='same', activation='relu',
input_shape=(28,28,1)))
model.add(layers.Conv2D (32, kernel_size=(3,3), padding='same', activation='relu'))
model.add(layers.Conv2D (32, kernel_size=(3,3), padding='same', activation='relu'))
model.add(layers.Flatten())
model.add(layers.Dense (128, activation='relu'))
model.add(layers.Dense (10, activation='softmax'))
```

```python
model.summary()
```

Each layer preserves the same spatial dimension, because padding compensates for the border loss.

Effect

*   Image size stays constant → 28×28 all through.

*   Flattened size is larger (25,088).

*   Parameters grow to 3.2M, since the dense layer receives a bigger input.

```python
model = models.Sequential()
model.add(layers.Conv2D (32, kernel_size=(3,3), padding='same', strides=(4,4), activation='relu',
input_shape=(28,28,1)))
model.add(layers.Conv2D (32, kernel_size=(3,3), padding='same', strides=(4,4), activation='relu'))
model.add(layers.Conv2D (32, kernel_size=(3,3), padding='same', strides=(4,4), activation='relu'))
model.add(layers.Flatten())
model.add(layers.Dense (128, activation='relu'))
model.add(layers.Dense (10, activation='softmax'))
```

```python
model.summary()
```

Each convolution now halves the resolution because stride = 2 causes the kernel to “jump” 2 pixels at a time.

Effect

*   Rapid downsampling: 28 → 14 → 7 → 4

*   Flattened size becomes very small (512).

*   Parameters drop drastically (~85k).

*   Model is lighter and faster, but captures coarser features.
